```python
def get_365_data(lat, lng):
    BASE_URL = f'https://api.weatherbit.io/v2.0/history/daily?lat={lat}&lon={lng}&start_date=2022-01-01&end_date=2022-12-31&key={API_KEY}'
    response = requests.get(f'{BASE_URL}')
    if response.status_code == 200:
        return response.json()
    else:
        return f'Request failed {response.status_code}'
```
