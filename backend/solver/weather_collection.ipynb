```python
def get_365_data(lat, lng):
    BASE_URL = f'https://api.openweathermap.org/data/2.5/onecall/timemachine?lat={lat}&lon={lng}&appid={API_KEY}'
    historical_weather_data = []
    t = (datetime.datetime(2022, 12, 31, 0, 0) - datetime.datetime(1970,1,1)).total_seconds()
    for i in range(365):
        timestamp = t - i * 86400  # Subtract i days in seconds
        response = requests.get(f'{BASE_URL}&dt={timestamp}')
        if response.status_code == 200:
            print(response.json())
            historical_weather_data.append(response.json())
        else:
            print(f'Request failed for timestamp {timestamp}. Error code: {response.status_code}')
```
