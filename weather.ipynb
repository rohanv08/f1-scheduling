```python
import pandas as pd
from datetime import datetime
import pickle
```

```python
#https://www.kaggle.com/datasets/hansukyang/temperature-history-of-1000-cities-1980-to-2020
#https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020

circuits = pd.read_csv("circuits.csv")
temp = pd.read_csv("temp.csv", skiprows = 1)
temp.drop(labels=[1,2,4,5,7,8,9], inplace = True)
```

```python
replacements = {'Al Daayen': 'Qatar', 
                'Anderstorp': 'Sweden', 
                'Castle Donington': 'United Kingdom',
                'Dijon': 'France',
                'Eastern Cape Province': 'South Africa',
                'Estoril': 'Portugal',
                'Heusden-Zolder': 'Belgium',
                'Hockenheim': 'Germany',
                'Imola': 'Italy',
                'Jerez de la Frontera': 'Spain',
                'Kent': 'London',
                'Le Castellet': 'France',
                'Le Mans': 'France',
                'Magny Cours': 'France',
                'Marina Bay': 'Singapore',
                'Midrand': 'South Africa',
                'Monte-Carlo': 'France',
                'Montmeló': 'Barcelona',
                'Monza': 'Italy',
                'Mugello': 'Florence',
                'New York State': 'New York',
                'Nürburg': 'Germany',
                'Oporto': 'Portugal',
                'Oyama': 'Japan',
                'Pescara': 'Italy',
                'Portimão': 'Portugal',
                'Reims': 'France',
                'Rouen': 'France',
                'Sakhir': 'Bahrain',
                'Silverstone': 'United Kingdom',
                'Sochi': 'Russia',
                'Spa': 'Belgium',
                'Spielberg': 'Austria',
                'Styria': 'Austria',
                'Suzuka': 'Japan',
                'Yeongam County': 'Seoul',
                'Zandvoort': 'Netherlands',
                'Clermont-Ferrand': 'France'}
```

```python
df = pd.DataFrame(columns = circuits.location.unique())
for col in df.columns:
  rep_col = col
  try:
    rep_col = replacements[col]
  except KeyError:
    pass

  try:
    df[col] = temp[rep_col]
  except KeyError:
    continue

for i in [0,3,6]:
  for col in df.columns:
    if df[col].isna().all():

      rep_col = col
      try:
        rep_col = replacements[col]
      except KeyError:
        pass
      matched = temp.columns[temp.loc[i] == rep_col]
      if len(matched) > 0:
        try:
          df[col] = temp[matched[0]]
        except KeyError:
          continue

df.index = temp["city"]
df.drop(['city_ascii','country','admin_name','datetime'], inplace = True)
df.index = pd.to_datetime(df.index)
```

```python
df.reset_index(inplace = True)
df.city = df.city.dt.dayofyear
df = df.astype(float).groupby("city").mean()
```

```python
dic = {}
for col in df.columns:
  dic[col] = df[col].values
pickle.dump(dic, open('weather.pkl', 'wb'))
```
